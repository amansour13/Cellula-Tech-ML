```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hotel.csv')
```

```python
df['date of reservation'] = pd.to_datetime(df['date of reservation'], errors='coerce')
```

```python
df['booking status'].value_counts()
```

```python
newdf = df
```

```python
newdf['booking status'] = newdf['booking status'].replace({'Not_Canceled':0, 'Canceled':1})
```

```python
# canceling trend along the reservation date
cancel_trend = newdf.groupby('date of reservation')['booking status'].mean()
```

```python
cancel_trend.plot(figsize=(12, 5), title="Cancellation Trend Over Reservation Dates")
plt.ylabel("Cancellation Rate")
plt.xlabel("Reservation Date")
plt.grid(True)
plt.style.use('seaborn-v0_8-colorblind')
plt.show()
```

```python
sns.boxplot(x='booking status', y='lead time', data=newdf, palette='mako')
plt.title('Lead Time vs Cancellation')
plt.xlabel('Canceled (0 = No, 1 = Yes)')
plt.ylabel('Lead Time (days)')
plt.show()
```

```python
cancel_rate = newdf.groupby(by='type of meal')['booking status'].mean().sort_values(ascending=False)
plt.figure(figsize=(6,6))
plt.pie(cancel_rate, labels=cancel_rate.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("mako", len(cancel_rate)))
plt.title("Cancellation Rate by Type of Meal")
plt.tight_layout()
plt.savefig('meal_types.png')
plt.show()
```

```python
cancel_rate = newdf.groupby(by='car parking space')['booking status'].mean().sort_values(ascending=False)
plt.figure(figsize=(5,5))
plt.pie(cancel_rate, labels=cancel_rate.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("mako", len(cancel_rate)))
plt.title("Cancellation Rate by car parking spaces\n(0 = no space, 1 = with space)")
plt.tight_layout()
plt.savefig('car_parking_space.png')
plt.show()
```

```python
most_canceled_room = newdf[newdf['booking status'] == 1]['room type'].value_counts().sort_values(ascending=False)
most_canceled_room.plot(kind='bar', title="Cancellation by Room Types", color=sns.color_palette("mako", len(cancel_rate)))
plt.ylabel("Cancelled")
plt.xlabel("Room types")
plt.grid(True)
plt.style.use('seaborn-v0_8-colorblind')
plt.savefig('room_types.png')
plt.show()
```

```python
cancel_rate = newdf.groupby(by='market segment type')['booking status'].mean().sort_values(ascending=False)
sns.barplot(x=cancel_rate.index, y=cancel_rate.values, palette="mako")
plt.title("Cancellation Rate by Market Segment Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('market_segment.png')
plt.show()
```

```python
request_cancel_rate = newdf.groupby('special requests')['booking status'].mean().sort_values(ascending=True).reset_index()

sns.barplot(x='special requests', y='booking status', data=request_cancel_rate, palette='mako')
plt.title("Cancellation Rate by Number of Special Requests")
plt.xlabel("Number of Special Requests")
plt.ylabel("Cancellation Rate")
plt.ylim(0, 1)
plt.savefig('special_requests.png')
plt.show()
```

```python
newdf['lead_bin'] = pd.cut(newdf['lead time'], bins=[0, 30, 60, 90, 120, 150, 200, 300, 400], right=False)

cancel_rate = newdf.groupby('lead_bin')['booking status'].mean().reset_index()

sns.barplot(x='lead_bin', y='booking status', data=cancel_rate, palette='mako')
plt.xticks(rotation=45)
plt.ylabel('Cancellation Rate')
plt.title('Cancellation Rate by Lead Time Bins')
plt.savefig('lead_time.png')
plt.show()
```
